* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
X = np.array([[1, 0, 1],
              [2, 0, 2],
              [3, 0, 3],
              [4, 4, 4]])

arr = X[((np.eye(*X.shape)).astype(bool)) & (X > 0)]
mult = np.cumprod(arr)
print(f'The result of multiplication is {mult[-1]}')# your code
```

```python
lst = [[1, 0, 1],
      [2, 0, 2],
      [3, 0, 3],
      [4, 4, 4]]
mult = 1
for i, elem in enumerate(lst):
    for j, el in enumerate(elem):
        if i == j and el != 0:
           mult *= el
print(f'The result of multiplication is {mult}')
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
x = np.array([2, 2, 2, 3, 3, 3, 5])

diff = np.diff(x)
idx = np.where(diff != 0)[0] + 1

values = np.split(x, idx)
values = np.array([v[0] for v in values])

counts = np.array([len(v) for v in np.split(x, idx)])

print(f'Values are: {values}, Repeat them {counts} times')
# your code
```

```python
x = [2, 2, 2, 3, 3, 3, 5]

values = []
counts = []

current_value = x[0]
current_count = 1

for elem in x[1:]:
    if elem == current_value:
        current_count += 1
    else:
        values.append(current_value)
        counts.append(current_count)
        current_value = elem
        current_count = 1


values.append(current_value)
counts.append(current_count)

print(f'Values are: {values}, Repeat them {counts} times')
```

scipy.spatial.distance.cdist на больших выборках работает быстрее

```python
import math

np.random.seed(17)
X = np.random.randint(1, 50, size=(1000, 2))
Y = np.random.randint(1, 50, size=(500, 2))

dist = []
for x1, x2 in X:
    row = []
    for y1, y2 in Y:
        dx = x1 - y1
        dy = x2 - y2
        row.append(math.sqrt(dx*dx + dy*dy))
    dist.append(row)

print(dist[0][:10])
```
