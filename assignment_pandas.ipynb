**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
dt = data.groupby(['occupation'])[['age', 'hours-per-week']].agg(['mean', 'min'])
```

```python
def filter_func(inpt: pd.DataFrame):
    inpt = inpt.groupby(['occupation'])[['age', 'hours-per-week']].agg(['mean', 'min', 'count'])
    inpt = inpt[
        (inpt[('age', 'mean')] <= 40) &
        (inpt[('hours-per-week', 'min')] > 5)
    ]
    inpt.columns = ['_'.join(col).strip() for col in inpt.columns.values]
    return inpt.drop(columns=list(set(inpt.columns) - {'age_count'})).rename(columns={'age_count': 'people'})

filter_func(data)
```
